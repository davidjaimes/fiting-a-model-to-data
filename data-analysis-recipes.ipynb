```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Exercise 1

```python
df = pd.read_csv('table1.dat', sep=' ', usecols=[1,2,3])
df = df[4:]
n = len(df)

Y = df.y.values.reshape((n, 1))
A = np.matrix([np.ones(n), df.x]).T
C = np.diag(pow(df.sig_y.values, 2))
inv = np.linalg.inv
CM1 = inv(A.T @ inv(C) @ A)
CM2 = A.T @ inv(C) @ Y
M = CM1 @ CM2
b, m = M.item(0), M.item(1)
em = np.sqrt(CM1.item(3))

fig, ax = plt.subplots()
xf = np.linspace(0, 300)
yf = b + m * xf
ax.plot(xf, yf, 'k')
ax.errorbar(df.x, df.y, df.sig_y, ls='none', marker='.', ms=8,
            zorder=3, label='Data', capsize=3, lw=0.5, color='k')
ax.text(100, 100, rf'$y = {b:.0f} + ({m:.2f} \pm {em:.2f})x$',
       size=16, color='k')
ax.set_xlabel('x-coordinate', size=16)
ax.set_ylabel('y-coordinate', size=16);
```

# Excericse 3

```python
df = pd.read_csv('table1.dat', sep=' ', usecols=[1,2,3])
df = df[4:]
n = len(df)

Y = df.y.values.reshape((n, 1))
A = np.matrix([np.ones(n), df.x, df.x ** 2]).T
C = np.diag(pow(df.sig_y.values, 2))
inv = np.linalg.inv
CM1 = inv(A.T @ inv(C) @ A)
CM2 = A.T @ inv(C) @ Y
M = CM1 @ CM2
b, m, q = M.item(0), M.item(1), M.item(2)

fig, ax = plt.subplots()
xf = np.linspace(0, 300)
yf = b + m * xf + q * xf ** 2
ax.plot(xf, yf, 'k')
ax.errorbar(df.x, df.y, df.sig_y, ls='none', marker='.', ms=8,
            zorder=3, label='Data', capsize=3, lw=0.5, color='k')
ax.text(100, 100, rf'$y = {b:.0f} + {m:.2f}x + {q:.4f}x^2$',
       size=16, color='k')
ax.set_xlabel('x-coordinate', size=16)
ax.set_ylabel('y-coordinate', size=16);
```
